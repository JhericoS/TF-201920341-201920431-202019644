```python
import numpy as np
def lectura(texto):
  rez = []
  with open(texto)as f:
    for line in f:
      rez.append(line.replace("\n", ""))

  return rez

Carriles_paralelos_A=list(reversed(lectura("carrilA.txt")))
Carriles_paralelos_B=lectura("carrilB.txt")
```

```python
def matr(m,n):
  maxtr=[]
  for x in range(1,m*n+1):
    if x%m==0 and not x==m*n:
      matrix2=[x+m-1]
    elif x==m*n: 
      maxtr.append([])
    elif m*n-m < x < m*n:
      matrix2=[x]
    else:
      matrix2=[x,x+m-1]
    maxtr.append(matrix2)
  maxtr.pop()
  return maxtr
        
nodo_cruces=matr(len(Carriles_paralelos_A),len(Carriles_paralelos_B))
print(len(Carriles_paralelos_A),len(Carriles_paralelos_B))
```

```python
for columnna in nodo_cruces:
    for elemnto in columnna:
        print(elemnto+771, end=' ')
    print()
```

```python
def matrix_cruces(G1,G2):
  #Carril paralelo a G1
  #Carril paralelo b G2
  cruces=[]
  for x in range(len(G1)):
    for j in range(len(G2)):
      cruces.append(G1[x]+" cruce con "+G2[j])
  return cruces

string_lista_carriles_cruces=matrix_cruces(Carriles_paralelos_B,Carriles_paralelos_A)#cuidado colocar el orden correcto No el lo mismo AxB q BxA
```

```python
for a in range(len(string_lista_carriles_cruces)):
    print(a+771,end=": ")
    print(string_lista_carriles_cruces[a])
```

```python
import graphviz  as gv
Verticeszzzzzzz=string_lista_carriles_cruces


lista_adyacencia=nodo_cruces

nombre_grafo=gv.Graph("los cruces en Argentina")


for a in range(len(Verticeszzzzzzz)):
  nombre_grafo.node(str(a),Verticeszzzzzzz[a])

for b in range(len(Verticeszzzzzzz)):
  for c in lista_adyacencia[b]:
    nombre_grafo.edge(str(b),str(c))#no acepta lista_adyacencia[c] suponmgo q acepta solo string

nombre_grafo 
```
